# 1. Les paramètres

## 1.1 Rappels

Quand vous devrez coder une fonction, vous commencerez par donner son **prototype**.

Les prototypes que vous devrez écrire donneront trois types d'informations :
1. le nom de la fonction
2. de 0 à N paramètres formels avec pour chacun un nom et un type
1. le type de son retour (ou `None` si la fonction ne retourne rien)

Le formalisme à respecter sera le suivant :

```python
nomDeFonction(nom1: type1, nom2: type2) -> typeRetour 
```

Pour utiliser une fonction il faut **l'appeler** depuis une autre partie du code.

- Si une fonction retourne quelque chose, il faut utiliser ce retour : dans une affectation, un affichage, etc.
- On passe à la fonction autant d'arguments qu'elle attend de paramètres : ces arguments peuvent être des valeurs, des noms de variable, des résultats d'opération, des retours de fonction.

**Important :** en Python, dans une fonction, la réaffectation d'un paramètre de cette fonction n'a aucun effet en dehors de la fonction elle-même, exactement comme pour une variable locale. Pour éviter toute confusion, **ne modifiez pas les paramètres** ! Utilisez-les seulement en lecture seule. **__Cependant ...__**

 ## 1.4 Passage d'une liste (`list`) en paramètre de fonction

Que fait la fonction ci-dessous ? Expliquez pourquoi.

````python
def swap ( indice1 : int , indice2 : int , liste : list ) -> None :
    if indice1 != indice2 and indice1 >= 0 and indice1 < len(liste) and indice2 >= 0 and indice2 < len(liste) :
        tmp = liste [indice1]
        liste[indice1] = liste[indice2]
        liste[indice2] = tmp
    else :
        print ("Echange impossible")
    return None
````

Elle ne fait rien car tout ce qui se passe dedans n'est stocké nulle part à la fin de son execution car elle n'a pas de return. Le seul résultat utile c'est lorsque l'on ne peut pas procéder à l'échange.

### 3.1.2 Plus de uns

Codez une fonction `plusDeUns()` dans laquelle vous tirez aléatoirement 0 ou 1 jusqu'à ce que la différence entre le nombre de 1 et le nombre de 0 obtenus soit `n`. Ce nombre `n` est passé en paramètre et le nombre de 1 est retourné par la fonction.

````python
def plusDeUns(n : int) -> int:
    nb1=0
    n1=0
    while n1!=n:
        nb=randint(0,1)
        if nb==0:
            n1+=-1
        else:
            n1+=1
            nb1+=1 
    return nb1
print(plusDeUns(50))
````

`t.getY()`
> Retourne la position verticale de la tortue `t`.

`t.left(angle)`
> Tourne la tortue `t` vers la gauche de `angle` en degrés.

`t.right(angle)`
> Tourne la tortue `t` vers la droite de `angle` en degrés.

`t.forward(distance)`
> Avance la tortue `t` dans sa direction selon `distance` en pixels.

`t.backward(distance)`
> Recule la tortue `t` dans sa direction selon `distance` en pixels.

Dans un second temps, vous remarquez que des lignes de code identiques ou très similaires se répètent 3 fois : les parties de transition d'une part, les dessins de figure d'autre part. Votre travail est de factoriser ces lignes de code en créant deux fonctions `avancer()`et `dessinerUnPolygone()` puis de simplifier la fonction `dessinerTroisPolygones()` en utilisant ces fonctions. Chacune de ces fonctions sera donc appelée 3 fois, avec les arguments adéquats : pour le carré, le triangle et l'hexagone. Vous pouvez également essayer de rassembler tous ces appels dans un parcours qui à chaque itération appelle une fois `avancer()`et `dessinerUnPolygone()`. 

````python
def avancer(color : str, turtle : Turtle)-> Turtle :
    turtle.penup()
    turtle.forward(200)
    turtle.pendown()
    turtle.setColor(color)
    return turtle
def dessinerUnPolygone(size : int, nbCote : int, turtle : Turtle, cv : Canvas) -> Turtle:
    turtle.left(180/nbCote)
    for i in range(nbCote):
        turtle.forward(size/4)
        turtle.left(180/nbCote*2)
        cv.displayPause(0.2) # cv est le canevas où la tortue est dessinée
    turtle.right(180/nbCote)
    return turtle

def dessinerTroisPolygones() -> None:
    # appel du constructeur pour créer l'objet cv
    cv = Canvas((800, 400))
    # appel du constructeur pour créer la tortue précisant canvas et center
    turtle = Turtle(cv, (50, 150))
    # affichage
    cv.displayPause(0.5)
    size = 400
    turtle=avancer(red,turtle)
    turtle=dessinerUnPolygone(size,4,turtle,cv)
    size = 600
    turtle=avancer(blue,turtle)
    turtle=dessinerUnPolygone(size,3,turtle,cv)
    size = 300
    turtle=avancer(green,turtle)
    turtle=dessinerUnPolygone(size,6,turtle,cv)

# test
dessinerTroisPolygones()
````
